→Total no of notification sent

```python
Total_no_of_Notifications = len(data)
```

In above code i am  storing total no of notification sent in Total_no_of_Notifications variable and  i am using len()
 function and passing data it will count
total no of rows and give it as length of data

Now lets print the no of notification sent

```python
print(Total_no_of_Notifications)
```

→Total clicks

```python
Total_clicks = data["Clicked"].sum()
print(Total_clicks) 
```

In the above i defined a variable name Total_clicks which stores the sum of clicks. In Clicked column it contain two type of values 0 and 1 mean not clicked and 1 mean clicked
as summing all will add only 1 and 0 have no impact so summing will indirectly give total no of clicked responce And after storing value i am printing it using print function

→Click through rate

It tells what percent of notifications got value=1 or we can say clicked 

```python
Click_through_rate = (Total_clicks /Total_no_of_Notifications)*100
print(Click_through_rate)
```

In the above code i am creating a variable name Click_through_rate in which i am storing (Total_clicks /Total_no_of_Notifications) AND MULTIPLYING IT WITH 100 TO CREATRE WHAT PERCENT OF NOTIFICATION IS IS CLICKED 

→Click by notification type

```python
Click_by_notification_type = data.groupby("Notification_type")["Clicked"].sum()
print(Click_by_notification_type)
```

In the above code i am just creating a variable name Click_by_notification_type in which i am calculating the no of clicked notifications for different type of 
notification separately
data.groupby("Notification_type") it groups all the type of notification from Notification_type , 
["Clicked"].sum() it sum all the clicked data mean add 1 ,  data.groupby("Notification_type")["Clicked"].sum() together in the code it will group all notification type and  calcluate total clicked value for each separately 

And in the last i am using print function to print output 

→Click by device type

```python
Click_by_device_type = data.groupby("Device_Type")["Clicked"].sum()
print(Click_by_device_type)
```

In the above code i am just creating a variable name Click_by_device_type in which i am storing total no of clicked by different device type accordingly 
data.groupby("Device_Type") - this will group different device separately 
["Clicked"].sum() - this will create sum of clicked notification or sum of 1 
data.groupby("Device_Type")["Clicked"].sum() - together it is calculating total no of clicked notification for differnt device type seperately

at end i used print to print the value inside Click_by_device_type

→Descriptive Stats

```python
data["Clicked"].describe()
```

In the above code in Clicked data which is 0 or 1 we calculate 
count,mean,std,min,25% ,50% ,75%,max using describe function 

→ Descriptive stats by category

```python
data.groupby("Notification_type")["Clicked"].describe()
```

In the above i am using describe on  Notification_type  Clicked which group different notification type Alert,Offer,Promotion,Reminder and describe its count mean 
std,min,25%,50%,75%,max on clicked for different notification separately

```python
data.groupby("Device_Type")["Clicked"].describe()
```

In the above i am using describe on  Device_Type  Clicked which group different Device_Type Mobile,Desktop,Tablet and describe its count mean 
std,min,25%,50%,75%,max on clicked for different device seperately

-->Bar diagram on basis of Clicks by notification type 

```python
plt.figure()
Click_by_notification_type.plot(kind='bar',color='yellow')
plt.title("Clicks by Notification Type")
plt.xlabel("Notification Type")
plt.ylabel("Number of Clicks") 
plt.grid() 
```

in the above code i am plotting a bar chart of Notification Type VS Number of Clicks us matplotlib library 
Click_by_notification_type.plot(kind='bar',color='yellow') ---  this will create a bar diagram for each notification type of yellow color 
plt.title("Clicks by Notification Type") --- This will create title for the Bar chart Clicks by Notification Type
plt.xlabel("Notification Type") --- This will create Label for x as Notification Type
plt.ylabel("Number of Clicks") --- This will create Label for y as Number of Clicks
plt.grid() --- This will create a grid like structure 

-->Bar diagram on basis of Clicks by Device type 

```python
plt.figure()
Click_by_device_type.plot(kind='bar',color='yellow')
plt.title("Clicks by Device Type")
plt.xlabel("Device Type")
plt.ylabel("Number of Clicks") 
plt.grid() 
```

in the above code i am plotting a bar chart of Device Type VS Number of Clicks us matplotlib library

plt.title("Clicks by Device Type") --- This will create title for the Bar chart Clicks by Device Type

plt.ylabel("Number of Clicks") --- This will create Label for y as Number of Clicks
plt.grid() --- This will create a grid like structure
